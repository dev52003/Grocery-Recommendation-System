```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("market_basket_transactions.csv")
```

```python
result = df.apply(lambda x: [item for item in x if pd.notna(item)], axis=1).tolist()
```

```python
result
```

```python
from apyori import apriori
rules = apriori(result,min_support=0.015,min_confidence=0.2,min_lift=1,min_length=2)
```

```python
rules_list = list(rules)
```

```python
def get_recommendations(product, rules_list):
    recommendations = []
    for rule in rules_list:
        for ordered_stat in rule.ordered_statistics:
            if product in ordered_stat.items_base:
                for item in ordered_stat.items_add:
                    recommendations.append((item, ordered_stat.confidence, ordered_stat.lift))
    recommendations.sort(key=lambda x: (x[2], x[1]), reverse=True)
    return recommendations
```

```python
product = "milk"
recommendations = get_recommendations(product, rules_list)
recommendations
```

```python
if recommendations:
    print(f"Recommendations for '{product}':")
    print(f"-{recommendations[0][0]}\n-{recommendations[1][0]}")
else:
    print(f"No recommendations found for '{product}'.")
```
